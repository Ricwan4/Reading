```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel('Emotion_Attention_Behavior.xlsx')

```

```python
# AB_Scale
df_ab = df.melt(
    id_vars=['Subject'],
    value_vars=['G-P_AB_Scale', 'G-N_AB_Scale', 'L-P_AB_Scale', 'L-N_AB_Scale'],
    var_name='Condition',
    value_name='AB_scale'
)
df_ab['QuestionType'] = df_ab['Condition'].apply(lambda x: 'Global' if 'G-' in x else 'Local')
df_ab['Emotion'] = df_ab['Condition'].apply(lambda x: 'Positive' if '-P_' in x else 'Negative')

# Accuracy
df_acc = df.melt(
    id_vars=['Subject'],
    value_vars=['G-P_Acc', 'G-N_Acc', 'L-P_Acc', 'L-N_Acc'],
    var_name='Condition',
    value_name='Accuracy'
)
df_acc['QuestionType'] = df_acc['Condition'].apply(lambda x: 'Global' if 'G-' in x else 'Local')
df_acc['Emotion'] = df_acc['Condition'].apply(lambda x: 'Positive' if '-P_' in x else 'Negative')

# RT
df_rt = df.melt(
    id_vars=['Subject'],
    value_vars=['G-P_RT', 'G-N_RT', 'L-P_RT', 'L-N_RT'],
    var_name='Condition',
    value_name='RT'
)
df_rt['QuestionType'] = df_rt['Condition'].apply(lambda x: 'Global' if 'G-' in x else 'Local')
df_rt['Emotion'] = df_rt['Condition'].apply(lambda x: 'Positive' if '-P_' in x else 'Negative')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# mean and error bar
fig, ax = plt.subplots(figsize=(10, 8))

sns.pointplot(
    data=df_ab,
    x='Emotion',
    y='AB_scale',  
    hue='QuestionType',
    dodge=0.3,
    capsize=0.1,
    err_kws={'linewidth': 3},
    palette={'Global': '#e41a1c', 'Local': '#377eb8'}, 
    markers='o'
)

# Overlay data point
sns.stripplot(
    data=df_ab,
    x='Emotion',
    y='AB_scale',
    hue='QuestionType',
    dodge=True,
    jitter=True,
    alpha=0.6,
    size=6,
    palette={'Global': '#fbb4ae', 'Local': '#b3cde3'}, 
    edgecolor='black',  
    linewidth=0
)

# Fix double legends
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[0:2], labels[0:2], title='Question Type',  fontsize=20, title_fontsize=20)

# Clean plot
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(1)

# Labels and Title
ax.set_xlabel('Emotion', fontsize=24)
ax.set_ylabel('G-L Scale', fontsize=24)
ax.tick_params(axis='both', which='major', labelsize=20)

plt.grid(False)
plt.tight_layout()
plt.show()
```
